```python
import numpy as np
import matplotlib.pyplot as plt
import netgraph
from netgraph import Graph
```

```python
matrix = np.array([
    [0,2,0,8,0],
    [4,0,0,0,6],
    [3,0,7,0,0],
    [4,0,6,0,0],
    [4,0,0,6,0]
])/10
```

```python
def get_layout(matrix, function_layout = netgraph.get_fruchterman_reingold_layout):
    sources, targets = np.where(matrix)
    weights = matrix[sources, targets]
    edges = list(zip(sources, targets))
    edge_labels = dict(zip(edges, weights))

    layout = function_layout(edges)

    fig, ax = plt.subplots()
    Graph(edges, edge_labels=edge_labels, edge_label_position=0.5, edge_label_fontdict=dict(size=10, fontweight='bold'),
        arrows=True, ax=ax, node_layout = layout, edge_layout='curved')
    plt.show()
    return layout

def print_coords_latex(layout):
    layout_to_latex = np.array([layout[i] for i in layout.keys()])

    # Recentering
    layout_to_latex = layout_to_latex - np.average(layout_to_latex, axis = 0)
    # Renormalizing
    layout_to_latex = layout_to_latex/np.min(np.linalg.norm(layout_to_latex, axis = 1)[0])
    # Rounding 
    layout_to_latex = np.round(layout_to_latex, 3).tolist()
    
    print(str(layout_to_latex).replace("]",")").replace("),", "),\n").replace("[","(")[1:-1])

def print_adjmatrix_latex(matrix):
    print(str(np.round(matrix,2).tolist()).replace("],", "],\n").replace("[", "{").replace("]", "}")[1:-1])

print_coords_latex(get_layout(matrix))
print()
print_adjmatrix_latex(matrix)
```
