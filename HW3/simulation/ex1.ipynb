```python
from itertools import permutations
from math import factorial
import numpy as np
import sympy as sp
n = 4
```

```python
card_permutations = [list(p) for p in permutations(range(n))]
def to_index(p):
    return card_permutations.index(p)

def shuffle_first(cards, position, i = 0):
    item = cards[i]
    shuffled = cards.copy()
    shuffled.remove(cards[i])
    shuffled.insert(position, item)
    return shuffled
```

```python
m = factorial(n)
matrix = np.zeros((m, m)).astype(int)
for i, cards in enumerate(card_permutations):
    for k in range(1):
        for position in range(n):
            if position == k:
                continue
            shuffled = shuffle_first(cards, position,k)
            j = to_index(shuffled)
            matrix[i,j] += 1

print(to_index([0,1,2,3]))
for k in range(1):
    for position in range(n):
        if position == k:
            continue
        shuffled = shuffle_first([0,1,2,3], position, k)
        j = to_index(shuffled)
        print(j)
```

```python
print(str(matrix.tolist()).replace("[","[").replace("],","],\n")[1:-1])
```

```python
error_list = []
for power in range(1,30):
    error = np.max(np.abs(np.linalg.matrix_power(matrix, power) - 1/24))
    error_list.append(error)
for i in range(4):
    error = error_list[i]
    print(f"e_{{{3*i+1}}} = "+ str(round(error_list[3*i] * 100,5)) + rf"\% & e_{{{3*i+2}}} = "+ str(round(error_list[3*i+1] * 100,5)) + rf"\% & e_{{{3*i+3}}} = " + str(round(error_list[3*i+2] * 100,5)) + r"\% \\")
```

```python
m = factorial(n)
matrix = np.zeros((m, m)).astype(int)
for i, cards in enumerate(card_permutations):
    for k in range(4):
        position = (k+1)%4
        shuffled = shuffle_first(cards, position,k)
        j = to_index(shuffled)
        matrix[i,j] += 1

print(to_index([0,1,2,3]))
for k in range(4):
    position = (k+1)%n
    shuffled = shuffle_first([0,1,2,3], position, k)
    j = to_index(shuffled)
    print(j, shuffled)
```
